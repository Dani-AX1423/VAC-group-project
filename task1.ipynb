```python
# Import pypdf, random and NLTK modules
```

```python
import random
import PyPDF2
import nltk
```

```python
# Download punkt package from NLTK
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Open pdf and get the complete text into a variable
```

```python
with open("assets/US_Declaration.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    text = ""

    for page in reader.pages:
        text += page.extract_text() + " "
```

```python
# Split the text into sentences
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
# Select a sentence using random and print it
```

```python
random_sentence = random.choice(sentences)
print('A random sentence:\n' + random_sentence + '\n')
```

```python
# Import word_tokenize from nltk.tokenize
```

```python
from nltk.tokenize import word_tokenize
```

```python
# Tokenizing the selected sentence and print the tokens
```

```python
tokens = word_tokenize(random_sentence)
print('The tokens:\n' + str(tokens))
```

```python
# Import PorterStemmer from nltk.stem
```

```python
from nltk.stem import PorterStemmer
```

```python
# Initialize and apply stemming
```

```python
stemmer = PorterStemmer()
stems = []
for word in tokens:
    stems += [stemmer.stem(word)]
```

```python
# Print the stem words
```

```python
print('The stem words:\n' + str(stems))
```
